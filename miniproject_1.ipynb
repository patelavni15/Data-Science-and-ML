```python
#What was the daily return average of a stock?
```

```python
stock_data_AAPL['Daily Return'] = stock_data_AAPL['Close'].pct_change() * 100
```

```python
stock_data_AAPL.head()
```

```python
average_daily_return = stock_data_AAPL['Daily Return'].mean()
average_daily_return
```

```python
#Add a new column â€˜Trendâ€™ whose values are based on the 'Daily Return'.
```

```python
stock_data_AAPL['Trend']= stock_data_AAPL['Daily Return']
```

```python
stock_data_AAPL
```

```python
threshold = 0  # Define threshold value (0 for neutral)
stock_data_AAPL['Trend'] = 'Neutral'  # Default trend is 'Neutral'
stock_data_AAPL.loc[stock_data_AAPL['Daily Return'] > threshold, 'Trend'] = 'Upward'
stock_data_AAPL.loc[stock_data_AAPL['Daily Return'] < -threshold, 'Trend'] = 'Downward'
print(stock_data_AAPL)
```

```python
#Visualize trend frequency through a Pie Chart.
```

```python
# Count the frequency of each trend
trend_counts = stock_data_AAPL['Trend'].value_counts()

# Create a pie chart to visualize the trend frequencies
plt.figure(figsize=(6, 6))
plt.pie(trend_counts, labels=trend_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Trend Frequency in Daily Returns')
plt.show()
```

```python
#What was the correlation between the daily returns of different stocks?
```

```python
import pandas as pd

stock_data_AAPL = pd.read_csv('E:/Edureka/AAPL.csv')
stock_data_GOOG = pd.read_csv('E:/Edureka/GOOG.csv')

stock_data_AAPL['Date'] = pd.to_datetime(stock_data_AAPL['Date'])
stock_data_AAPL.set_index('Date', inplace=True)

stock_data_GOOG['Date'] = pd.to_datetime(stock_data_GOOG['Date'])
stock_data_GOOG.set_index('Date', inplace=True)

stock_data_AAPL['Daily Return'] = stock_data_AAPL['Close'].pct_change() * 100
stock_data_GOOG['Daily Return'] = stock_data_GOOG['Close'].pct_change() * 100

merged_data = pd.concat([stock_data_AAPL['Daily Return'], stock_data_GOOG['Daily Return']], axis=1)

correlation_matrix = merged_data.corr()

print(correlation_matrix)
```
