```python
#Problem Statement 3: Write a Python program to create a series from 1 to 1000 and select only numbers divisible by 7 and 17.
```

```python
# Create a Pandas Series with numbers from 1 to 1000
series_data = pd.Series(range(1, 1001))

# Select numbers that are divisible by both 7 and 17
divisible_by_7_and_17 = series_data[(series_data % 7 == 0) & (series_data % 17 == 0)]

# Display the selected numbers
print(divisible_by_7_and_17)
```

```python
from google.colab import files


uploaded = files.upload()
```

```python
import pandas as pd
import io

cereal_df = pd.read_csv(io.BytesIO(uploaded['cereal.csv']))
cereal_df
```

```python
#Plot ratings of different types of manufacturers.
import pandas as pd
import matplotlib.pyplot as plt
import io



# Read the CSV file into a Pandas DataFrame
cereal_df = pd.read_csv(io.BytesIO(uploaded['cereal.csv']))
cereal_df

# Group the data by manufacturer (MFR) and calculate the mean rating for each manufacturer
manufacturer_ratings = cereal_df.groupby('mfr')['rating'].mean()

# Create a bar plot to visualize the ratings for each manufacturer
plt.figure(figsize=(10, 6))
manufacturer_ratings.plot(kind='bar')
plt.title('Average Ratings by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Average Rating')
plt.xticks(rotation=0)  # Rotate x-axis labels for readability

# Show the plot
plt.show()
```

```python
#3. Use xticks range form 0-100.
```

```python

# Create a bar plot to visualize the ratings for each manufacturer
plt.figure(figsize=(10, 6))
manufacturer_ratings.plot(kind='bar')
plt.title('Average Ratings by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Average Rating')

# Set the range of x-axis ticks from 0 to 100
plt.xticks(range(len(manufacturer_ratings)), manufacturer_ratings.index, rotation=0)
plt.ylim(0, 100)  # Set the y-axis limit to 0-100

# Show the plot
plt.show()
```

```python
#4. Change the style of the graph to seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the Seaborn style
sns.set_style("whitegrid")

# Create a bar plot to visualize the ratings for each manufacturer
plt.figure(figsize=(10, 6))
sns.barplot(x=manufacturer_ratings.index, y=manufacturer_ratings.values)
plt.title('Average Ratings by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Average Rating')

# Set the range of y-axis from 0 to 100
plt.ylim(0, 100)

# Show the plot
plt.show()
```
